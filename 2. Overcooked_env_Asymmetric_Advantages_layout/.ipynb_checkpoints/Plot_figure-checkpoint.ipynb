```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

len_plot = int(1499)
len_window = 5
```

```python
# print('\n---------------------------------------- compare (sparse) ----------------------------------------\n')

data_caci_return_sparse = np.load('./data_tmp/data_caci_return_sparse.npy')
data_cippo_return_sparse = np.load('./data_tmp/data_cippo_return_sparse.npy')
data_ppobc_return_sparse = np.load('./data_tmp/data_ppobc_return_sparse.npy')

fig, ax = plt.subplots(1, 1, figsize=(8, 5), dpi=70)
plt.plot(data_caci_return_sparse[0:len_plot], '--',  markerfacecolor='none', label = 'CaCi')
plt.plot(data_cippo_return_sparse[0:len_plot], '-.', markerfacecolor='none', label = 'CI-PPO')
plt.plot(data_ppobc_return_sparse[0:len_plot], ':',  markerfacecolor='none', label = 'PPO-BC')
plt.xlabel('Epoch', fontsize = 16)
plt.ylabel('Sparse return (compare)', fontsize = 16)
plt.legend(fontsize = 16, loc='best')
plt.tick_params(labelsize=16)
fig.subplots_adjust(right=0.96, top=0.95, hspace=1, wspace=1)
fig.subplots_adjust(left=0.15, bottom=0.12)
# plt.savefig('./figs/compare_avg_return.pdf', format='pdf')
```

```python
# print('\n---------------------------------------- compare (shaped) ----------------------------------------\n')

data_caci_return_shaped = np.load('./data_tmp/data_caci_return_shaped.npy')
data_cippo_return_shaped = np.load('./data_tmp/data_cippo_return_shaped.npy')
data_ppobc_return_shaped = np.load('./data_tmp/data_ppobc_return_shaped.npy')


fig, ax = plt.subplots(1, 1, figsize=(8, 5), dpi=70)
plt.plot(data_caci_return_shaped[0:len_plot], '-',  markerfacecolor='none', label = 'Caci')
plt.plot(data_cippo_return_shaped[0:len_plot], '-.', markerfacecolor='none', label = 'CI-PPO')
plt.plot(data_ppobc_return_shaped[0:len_plot], ':',  markerfacecolor='none', label = 'PPO-BC')
plt.xlabel('Epoch', fontsize = 16)
plt.ylabel('Shaped return (compare)', fontsize = 16)
plt.legend(fontsize = 16, loc='best')
plt.tick_params(labelsize=16)
fig.subplots_adjust(right=0.96, top=0.95, hspace=1, wspace=1)
fig.subplots_adjust(left=0.15, bottom=0.12)
# plt.savefig('./figs/compare_avg_return.pdf', format='pdf')
```

```python
# print('\n---------------------------------------- ppobc ----------------------------------------\n')

data_ppobc_return_sparse = np.load('./data_tmp/data_ppobc_return_sparse.npy')
data_ppobc_return_shaped = np.load('./data_tmp/data_ppobc_return_shaped.npy')
data_ppobc_return_env    = np.load('./data_tmp/data_ppobc_return_env.npy')


fig, ax = plt.subplots(1, 1, figsize=(8, 5), dpi=70)
plt.plot(data_ppobc_return_sparse, '-',  markerfacecolor='none', label = 'Reward sparse')
plt.plot(data_ppobc_return_shaped, '-.', markerfacecolor='none', label = 'Reward shaped')
plt.plot(data_ppobc_return_env,    '--', markerfacecolor='none', label = 'Reward env')
plt.xlabel('Epoch', fontsize = 16)
plt.ylabel('return (ppobc)', fontsize = 16)
# plt.legend(fontsize = 16, loc='best')
plt.tick_params(labelsize=16)
fig.subplots_adjust(right=0.96, top=0.95, hspace=1, wspace=1)
fig.subplots_adjust(left=0.15, bottom=0.12)
plt.savefig('./figs/ppobc_avg_return.pdf', format='pdf')
```

```python
import matplotlib.pyplot as plt
import numpy as np

len_plot = int(149)


# 
```

```python
data_nobc_Return_mean_both   = np.load('./data_tmp/data_nobc_Reward_episode_env.npy')
data_nobc_Return_mean_sparse = np.load('./data_tmp/data_nobc_Reward_episode_sparse.npy')
data_nobc_Return_mean_shaped = np.load('./data_tmp/data_nobc_Reward_episode_shaped.npy')

fig, ax = plt.subplots(1, 1, figsize=(9, 6.5), dpi=50)
plt.plot(data_nobc_Return_mean_sparse[0:len_plot], '-',  markerfacecolor='none', label = 'return_mean sparse')
plt.plot(data_nobc_Return_mean_shaped[0:len_plot], '-.', markerfacecolor='none', label = 'return_mean shaped')
plt.plot(data_nobc_Return_mean_both[0:len_plot],   '--', markerfacecolor='none', label = 'return_mean env')
plt.xlabel('Epoch')
plt.ylabel('Return_mean')
plt.legend(fontsize=16, loc='best')
fig.subplots_adjust(right=0.96, top=0.95, hspace=1, wspace=1)
fig.subplots_adjust(left=0.15, bottom=0.12)
plt.savefig('./figs/nobc_avg_return.pdf', format='pdf')  
```

```python
# print('\n---------------------------------------- ppobc ----------------------------------------\n')

data_ppobc_Return_episode_shaped = np.load('./data_tmp/data_ppobc_Reward_episode_shaped.npy')
data_ppobc_Return_episode_sparse = np.load('./data_tmp/data_ppobc_Reward_episode_sparse.npy')
data_ppobc_Return_episode_env    = np.load('./data_tmp/data_ppobc_Reward_episode_env.npy')

fig, ax = plt.subplots(1, 1, figsize=(9, 6.5), dpi=50)
plt.plot(data_ppobc_Return_episode_shaped, '-',  markerfacecolor='none', label = 'Reward shaped')
plt.plot(data_ppobc_Return_episode_sparse, '-.', markerfacecolor='none', label = 'Reward sparse')
plt.plot(data_ppobc_Return_episode_env,    '--', markerfacecolor='none', label = 'Reward env')
plt.xlabel('Epoch', fontsize = 16)
plt.ylabel('Return (ppobc)', fontsize = 16)
plt.legend(fontsize = 16, loc='best')
fig.subplots_adjust(right=0.96, top=0.95, hspace=1, wspace=1)
fig.subplots_adjust(left=0.15, bottom=0.12)
plt.savefig('./figs/ppobc_avg_return.pdf', format='pdf')
```

```python
# print('\n---------------------------------------- cimar ----------------------------------------\n')

data_cimar_Return_episode_shaped = np.load('./data_tmp/data_cimar_Reward_episode_shaped.npy')
data_cimar_Return_episode_sparse = np.load('./data_tmp/data_cimar_Reward_episode_sparse.npy')
data_cimar_Return_episode_env    = np.load('./data_tmp/data_cimar_Reward_episode_env.npy')

fig, ax = plt.subplots(1, 1, figsize=(9, 6.5), dpi=50)
plt.plot(data_cimar_Return_episode_shaped, '-',  markerfacecolor='none', label = 'Reward shaped')
plt.plot(data_cimar_Return_episode_sparse, '-.', markerfacecolor='none', label = 'Reward sparse')
plt.plot(data_cimar_Return_episode_env,    '--', markerfacecolor='none', label = 'Reward env')
plt.xlabel('Epoch', fontsize = 16)
plt.ylabel('Reward (cimar)', fontsize = 16)
plt.legend(fontsize = 16, loc='best')
fig.subplots_adjust(right=0.96, top=0.95, hspace=1, wspace=1)
fig.subplots_adjust(left=0.15, bottom=0.12)
plt.savefig('./figs/cimar_avg_return.pdf', format='pdf')
```

```python
# print('\n---------------------------------------- compare ----------------------------------------\n')

fig, ax = plt.subplots(1, 1, figsize=(9, 6.5), dpi=50)
plt.plot(data_nobc_Return_mean_sparse[0:len_plot],      '-', markerfacecolor='none', label = 'ppono')
plt.plot(data_ppobc_Return_episode_sparse[0:len_plot], '-.', markerfacecolor='none', label = 'ppobc')
plt.plot(data_cimar_Return_episode_sparse[0:len_plot], '-.', markerfacecolor='none', label = 'cimar')
plt.xlabel('Epoch', fontsize = 16)
plt.ylabel('Return (compare)', fontsize = 16)
plt.legend(fontsize = 16, loc='best')
fig.subplots_adjust(right=0.96, top=0.95, hspace=1, wspace=1)
fig.subplots_adjust(left=0.15, bottom=0.12)
plt.savefig('./figs/compare_avg_return.pdf', format='pdf')
```
